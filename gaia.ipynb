```python
from astropy.table import Table, vstack, join, hstack, Column
import astropy.units as u
from astropy.coordinates import SkyCoord
from astroquery.gaia import Gaia
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
coord_a = SkyCoord(ra=222.50, dec=-7.10, unit=(u.degree, u.degree), frame='icrs')
radius = u.Quantity(1.5, u.deg)
j_a = Gaia.cone_search_async(coord_a, radius)
pal5_a = j_a.get_results()
#Reading in the pal5_a coordinates from gaiaDR2
```

```python
pm = (pal5_a['pmra']< -2.11) & (pal5_a['pmra']> -2.48) & (pal5_a['pmdec']< -2.08) & (pal5_a['pmdec']> -2.42)
pal5_a[pm]
#Applying a mask on the proper motion of the stars in this table. 20 stars were in this range.
```

```python
coord_b = SkyCoord(ra=240.50, dec=6.16, unit=(u.degree, u.degree), frame='icrs')
radius = u.Quantity(1.5, u.deg)
j_b = Gaia.cone_search_async(coord_b, radius)
pal5_b = j_b.get_results()
#Reading in the pal5_b coordinates from gaiaDR2
```

```python
pm1 = (pal5_b['pmra']< -2.11) & (pal5_b['pmra']> -2.48) & (pal5_b['pmdec']< -2.08) & (pal5_b['pmdec']> -2.42)
pal5_b[pm1]
#Applying a mask on the proper motion of the stars in this table. 53 stars were in this range.
```

```python
coord = SkyCoord(ra=229.02, dec=-0.12, unit=(u.degree, u.degree), frame='icrs')
radius = u.Quantity(1.5, u.deg)
j = Gaia.cone_search_async(coord, radius)
pal5_cluster = j.get_results()
#Reading in the pal5 cluster coordinates from gaiaDR2
```

```python
pal5_cluster
#I found 38,386 stars in this area
```

```python
plt.scatter(pal5_a[pm]['ra'],pal5_a[pm]['dec'],color='b')
plt.scatter(pal5_cluster['ra'],pal5_cluster['dec'],color='r')
plt.scatter(pal5_b[pm1]['ra'],pal5_b[pm1]['dec'], color='g')
plt.xlabel('RA')
plt.ylabel('DEC')
plt.show()
#This is a plot of where the pal5 stream stars that fit the proper motion mask are situated compared to the original cluster
```

```python
pal5a_stream = pal5_a[pm]
pal5a_stream.keep_columns(['ra','dec'])
pal5b_stream = pal5_b[pm1]
pal5b_stream.keep_columns(['ra','dec'])
#Getting rid of all the columns accept for RA and DEC.
```

```python
Table.write(pal5a_stream,'pal5a_stream.csv',format='ascii',overwrite='True',delimiter=',')
Table.write(pal5b_stream,'pal5b_stream.csv',format='ascii',overwrite='True',delimiter=',')
#Writing these tables into files so I can cross match them 2MASS data.
```
